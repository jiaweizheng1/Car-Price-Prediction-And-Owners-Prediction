Import libraries and car dataset, and print first five rows.

```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import seaborn as sns

cardata = pd.read_csv("car data.csv")

print(cardata.head())
```

Print shape/dimensions of car data.

```python
print(cardata.shape)
```

Print data type of variables in car data.

```python
print(cardata.info())
```

Print statistical report of variables in car data.

```python
print(cardata.describe())
```

Car data preprocessing

```python
cardata['Age'] = 2023 - cardata['Year']
cardata.drop(labels='Year', axis=1, inplace = True)

cardata.rename(columns = {'Owner':'Past_Owners'}, inplace = True)

cardata.drop(labels='Car_Name', axis= 1, inplace = True)

cardata = pd.get_dummies(data = cardata, drop_first=True)

print(cardata.head())
```

Heatmap analysis of correlation between variables in car data.

```python
sns.heatmap(cardata.corr(), annot=True, cmap="coolwarm", vmin=-1, vmax=1, center= 0,
            linewidths=1, linecolor='black')
plt.show()
```
