```python
import pandas as pd
import numpy as np
import seaborn as sns
```

```python
df = pd.read_csv('CAR DETAILS FROM CAR DEKHO.csv')
```

```python
df = df.set_index('name')
```

```python
catg = df[['year','fuel','seller_type','transmission','owner']]
for i in catg.columns:
    print(catg[i].value_counts())
```

```python
df['owner'] = df['owner'].apply(lambda x:0 if x == "First Owner" else 1).astype(int) # 0 First owner means 
#new car 1 second and above means used. This definition seems clear and more simplified
```

```python
sorted(df['year'].unique())
```

```python
df['year'] = df['year'].apply(lambda x:0 if x<2000 else 
                              (1 if x>=2000 and x<=2005 else(
                               2 if x>=2006 and x<=2010 else(
                               3 if x>=2011 and x<=2015 else 4)))).astype(str)
#0 Any car from before 2000 1 2000-2005 2 2006-2010 3 2011-2015 4 Any car after 2015
```

```python
df
```

```python
df['fuel'] = df['fuel'].apply(lambda x:'Petrol' if x == 'Petrol' else ('Diesel' if x == 'Diesel' else 'Other'))
```

```python
df = pd.get_dummies(df, columns = ['year','fuel','seller_type','transmission'])
```

```python
df
```

```python
df = df.drop(columns = ['year_0','fuel_Other','seller_type_Trustmark Dealer','transmission_Automatic'])
```

```python
owner = df.pop('owner')
df.insert(0,'owner',owner)
```

```python
df.to_csv('Cars.csv')
```
